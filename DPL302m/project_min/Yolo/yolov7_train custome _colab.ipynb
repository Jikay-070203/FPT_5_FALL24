```python
import torch
from ultralytics import YOLO
from pathlib import Path
import numpy as np

# --- Cấu hình ---
model_path = '/content/drive/MyDrive/yolov7_train/yolov7/runs/train/exp3/weights/best.pt'  # Thay bằng đường dẫn đến file .pt của mô hình đã huấn luyện
data_path = 'data/african-wildlife.yaml'       # Đường dẫn đến file .yaml của tập dữ liệu

# --- Hàm tính toán Mean center difference ---
def calculate_mean_center_difference(pred_boxes, true_boxes):
    """
    Tính toán khoảng cách trung bình giữa tâm của các bounding box dự đoán và ground truth.
    """
    center_differences = []
    for pred_box, true_box in zip(pred_boxes, true_boxes):
        pred_center = ((pred_box[0] + pred_box[2]) / 2, (pred_box[1] + pred_box[3]) / 2)
        true_center = ((true_box[0] + true_box[2]) / 2, (true_box[1] + true_box[3]) / 2)
        distance = np.sqrt(((pred_center[0] - true_center[0]) ** 2) + ((pred_center[1] - true_center[1]) ** 2))
        center_differences.append(distance)
    return np.mean(center_differences)

# --- Hàm tính toán Mean difference bounding box size ---
def calculate_mean_bbox_size_difference(pred_boxes, true_boxes):
    """
    Tính toán chênh lệch kích thước trung bình giữa các bounding box dự đoán và ground truth.
    """
    size_differences = []
    for pred_box, true_box in zip(pred_boxes, true_boxes):
        pred_size = (pred_box[2] - pred_box[0]) * (pred_box[3] - pred_box[1])
        true_size = (true_box[2] - true_box[0]) * (true_box[3] - true_box[1])
        difference = abs(pred_size - true_size)
        size_differences.append(difference)
    return np.mean(size_differences)

# --- Tải mô hình YOLOv7 ---
model = YOLO(model_path)

# --- Đánh giá mô hình ---
results = model.val(data=data_path, save_json=True, save_txt=True)

# --- Trích xuất bounding boxes từ kết quả ---
pred_boxes = []
true_boxes = []
for result in results.pred:
    for *box, conf, cls in result:
        pred_boxes.append(box)
for result in results.truth:
    for *box, cls in result:
        true_boxes.append(box)

# --- Tính toán Mean center difference và Mean difference bounding box size ---
mean_center_difference = calculate_mean_center_difference(pred_boxes, true_boxes)
mean_bbox_size_difference = calculate_mean_bbox_size_difference(pred_boxes, true_boxes)

# --- In kết quả đánh giá ---
print(f'Precision: {results.box.p:.3f}')
print(f'Recall: {results.box.r:.3f}')
print(f'F1-score: {results.box.f1:.3f}')
print(f'Accuracy: {results.box.acc:.3f}')
print(f'Mean center difference: {mean_center_difference:.3f}')
print(f'Mean difference bounding box size: {mean_bbox_size_difference:.3f}')
```
