### Explain the parameters
+ verified: Kết quả Boolean (đúng hoặc sai) xem hai ảnh có phải là cùng một người không.
+ distance: Khoảng cách biểu thị độ tương đồng.(càng nhỏ - càng tương đồng)
+ threshold: Ngưỡng để xác định một sự khớp.(Ngưỡng - để xác định )
+ Mô hình sử dụng, ví dụ như "Facenet", "VGG-Face", hoặc "ArcFace".
+ similarity_metric: Phương pháp so sánh, ví dụ cosine similarity

```r
from flask import Flask, request, jsonify
from flask_ngrok import run_with_ngrok
from deepface import DeepFace
from PIL import Image
from io import BytesIO
import base64

app = Flask(__name__)
run_with_ngrok(app)  # This will open a public URL for the app

# Define the verification function
def verify_faces(img1, img2, model_name="Facenet"):
    try:
        result = DeepFace.verify(img1, img2, model_name=model_name)
        return {"verified": result["verified"]}
    except Exception as e:
        return {"error": str(e)}

@app.route('/verify', methods=['POST'])
def verify():
    data = request.get_json()
    img1_data = data['image1']
    img2_data = data['image2']
    model_name = data.get('model_name', 'Facenet')  # Default to Facenet

    # Decode base64 images
    img1 = Image.open(BytesIO(base64.b64decode(img1_data)))
    img2 = Image.open(BytesIO(base64.b64decode(img2_data)))

    # Run verification
    result = verify_faces(img1, img2, model_name)
    return jsonify(result)

# Start the Flask app with ngrok
app.run()
```
